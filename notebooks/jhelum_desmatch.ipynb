```python
# Third-party
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from astropy.table import Table
from astropy.io import fits
import astropy.units as u
```

```python
t = Table(fits.getdata('../data/fits/jhelum_des.fits'))
```

```python
t.colnames
```

```python
for band in ['g', 'r', 'i']:
    t.rename_column('des_dr1_main.wavg_mag_psf_{}'.format(band), band)
    t.rename_column('des_dr1_main.wavg_magerr_psf_{}'.format(band), '{}_error'.format(band))
    t.rename_column('des_dr1_main.wavg_mag_psf_{}_dered'.format(band), '{}0'.format(band))

    t[band] *= u.mag
    t['{}_error'.format(band)] *= u.mag
    t['{}0'.format(band)] *= u.mag
```

```python
t.colnames
```

```python
t.write('../data/jhelum_des.fits', overwrite=True)
```
